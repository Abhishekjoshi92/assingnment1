6. Write a function to demonstrate the behavior of mutable and immutable arguments.



```python
# Immutable example with integer
def change_number(num):
    num = num + 5
    print("Inside function (number):", num)

# Mutable example with list
def change_list(my_list):
    my_list[0] = my_list[0] + 5
    print("Inside function (list):", my_list)

# Main part
number = 10
numbers = [10, 20, 30]

print("Before function call:")
print("number:", number)
print("numbers:", numbers)

change_number(number)
change_list(numbers)

print("After function call:")
print("number:", number)     # Will stay the same
print("numbers:", numbers)   # Will be changed
```

10. Write code to demonstrate type casting with list elements.

```python
# Original list with mixed types
mixed_list = [1, '2', 3.5, '4.7', 5]

# Convert elements to integers
int_list = [int(item) if type(item) == str and item.isdigit() else item for item in mixed_list]

# Convert elements to floats
float_list = [float(item) for item in mixed_list]

# Convert elements to strings
str_list = [str(item) for item in mixed_list]

# Print results
print("Original list:", mixed_list)
print("List as integers:", int_list)
print("List as floats:", float_list)
print("List as strings:", str_list)
```

13. Write a Python program to find the sum of all even numbers between 1 and 50.

```python
# start the sum
sum_even = 0

# Loop through numbers from 1 to 999
for num in range(1, 1000):
    if num % 2 == 0:  # Check if the number is even
        sum_even += num  # Add even number to the sum

# Print the result
print("Sum of all even numbers between 1 and 999:", sum_even)
```

14. Write a program to reverse a string using a while loop.

```python
# Take input from the user
input_string = input("Enter a string: ")

# Initialize an empty string for reversed output
reversed_string = ""

# Initialize a variable for the last index
index = len(input_string) - 1

# Use a while loop to reverse the string
while index >= 0:
    reversed_string += input_string[index]
    index -= 1

# Print the reversed string
print("Reversed string:", reversed_string)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
# Take input from the user
num = int(input("Enter a number: "))

# Initialize factorial as 1
factorial = 1

# Use a while loop to calculate the factorial
while num > 0:
    factorial *= num  # Multiply the current factorial by num
    num -= 1  # Decrease num by 1

# Print the result
print("Factorial is:", factorial)
```
